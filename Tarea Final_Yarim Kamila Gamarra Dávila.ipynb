## III. Funciones

```python
def calcular_craest(media_personal, media_curso, desviacion_curso):
    '''
    Objetivo:
        - Calcular el craest de un estudiante dado su media personal, la media del curso y la desviación estándar del curso
    
    Input:
        - media_personal: promedio del 0-20 del estuadiante
        - media_curso: promedio del  0-20 del salón
        - desviacion_curso: dispersión de notas del salón

    Output:
        - craest calculado
    '''
    craest = (((media_personal - media_curso)* 10)/desviacion_curso) + 50
    return craest
```

```python
resultado=calcular_craest(14,15,3.5)
print(resultado)
```
